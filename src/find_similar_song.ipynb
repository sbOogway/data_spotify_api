```python
import pandas as pd
import os

file_path = os.path.join("..", "data", "guess_tha_song.songs.features.json")

df = pd.read_json(file_path)

df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cdist

features = ["acousticness", "energy", "liveness", "speechiness", "danceability", "instrumentalness", "loudness", "valence"]
song = ["name", "artists"]

Features =  df[features].to_numpy()


df["distances"] = cdist(df[df["name"] == "Uptown Girl"][features], Features, metric="euclidean")[0]


df.sort_values(by="distances", inplace=True)

df
```
